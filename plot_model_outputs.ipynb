```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
d = '/home/campratt/Desktop/RESEARCH/SZU/outputs_evaluate/'

IDs = [0,1]

for idx, ID in enumerate(IDs):
    y_pred = np.load(os.path.join(d + f'y_{ID}_pred.npy'))[0]
    y_true = np.load(os.path.join(d + f'y_{ID}_true.npy'))[0][0]

    print(y_pred.shape, y_true.shape)  

    y_pred -= np.median(y_pred.flatten())
    y_true -= np.median(y_true.flatten())


    vmin = -5
    vmax = 20

    plt.imshow(y_pred, vmin=vmin, vmax=vmax)
    plt.colorbar()
    plt.show()

    plt.imshow(y_true, vmin=vmin,vmax=vmax)
    plt.colorbar()
    plt.show()
```
